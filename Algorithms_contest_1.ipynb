# A. Минимальный делитель числа

image.png

```python
n = int(input())

d = 2
ans = []
while d * d <= n:
  if n % d == 0:
    ans.append(d)
    n = n // d
  else:
    d += 1
if n != 1:
  ans.append(n)

print(min(ans))
```

# B. Наибольший общий делитель

image.png

```python
n = int(input())
m = int(input())

def gcd(a: int, b: int) -> int:
  while a != 0 and b != 0:
    if a > b:
      a = a % b
    else:
      b = b % a
  return a + b

print(gcd(n, m))
```

# C. Наименьшее общее кратное

photo_2024-11-26_02-53-00.jpg

```python
n = int(input())
m = int(input())

def gcd(a: int, b: int) -> int:
  while a != 0 and b != 0:
    if a > b:
      a = a % b
    else:
      b = b % a
  return a + b

def lcm(a: int, b: int) -> int:
  return a // gcd(a, b) * b

print(lcm(n, m))
```

# E. Сильно составные числа

image.png

```python
n = int(input())
sieve = [1] * (n + 1)
primes = []


for i in range(2, int(n ** 0.5) + 1):
    if sieve[i]:
        for j in range(i * i, n + 1, i):
            sieve[j] = 0


for i in range(2, n + 1):
    if sieve[i]:
        primes.append(i)


result = []
for num in range(2, n + 1):
    count = 0
    temp_num = num

    for p in primes:
        if p * p > temp_num:  # No need to check beyond sqrt of the number
            break
        if temp_num % p == 0:
            count += 1
            while temp_num % p == 0:
                temp_num //= p
        if count >= 3:  # Early exit if we find at least 3 distinct primes
            break
    if temp_num > 1:
        count += 1  # There is a prime factor greater than sqrt(n)

    if count >= 3:
        result.append(num)

print(*result)
```

# L. Факториал по модулю

image.png

```python
n = int(input())

modul = 10**6 + 3

if n >= modul:
    print(0)
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial = (factorial * i) % modul
    print(factorial)
```
